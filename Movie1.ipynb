## Working on Credits Dataset

```python
df1 = pd.read_csv("movies_metadata.csv")
df2 = pd.read_csv("credits.csv")
df2.head()
```

```python
df1.head()
```

```python
df1['year'] = pd.to_datetime(df1['release_date'], errors='coerce' )
```

```python
df1['year'] = df1['year'].dt.year
```

```python
df1['year'].value_counts().sort_index()
```

```python
new_df = df1.loc[df1.year == 2017,['genres','id','title','year']]
```

```python
new_df
```

```python
new_df['id'] = new_df['id'].astype(int)
```

```python
new_df.dtypes
```

```python
data = pd.merge(new_df, df2, on='id')
```

```python
data.head()
```

```python
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
data['genres_list'] = data['genres'].map(lambda x: make_genresList(x))
```

```python
data['genres_list']
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
data['actor_2_name'] = data['cast'].map(lambda x: get_actor2(x))
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
data['actor_3_name'] = data['cast'].map(lambda x: get_actor3(x))
```

```python
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data['director_name'] = data['crew'].map(lambda x: get_directors(x))
```

```python
movie = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title']]
```

```python
movie
```

```python
movie = movie.dropna(how='any')
```

```python
movie.isna().sum()
```

```python
movie = movie.rename(columns={'genres_list':'genres'})
movie = movie.rename(columns={'title':'movie_title'})
movie['movie_title'] = movie['movie_title'].str.lower()
```

```python
movie['comb'] = movie['actor_1_name'] + ' ' + movie['actor_2_name'] + ' '+ movie['actor_3_name'] + ' '+ movie['director_name'] +' ' + movie['genres']
```

```python
old = pd.read_csv('gbc1.csv')
```

```python
old['comb'] = old['actor_1_name'] + ' ' + old['actor_2_name'] + ' '+ old['actor_3_name'] + ' '+ old['director_name'] +' ' + old['genres']
```

```python
new = old.append(movie)
```

```python
new.shape
```

```python
new.drop_duplicates(subset ="movie_title", keep = 'last', inplace = True)
```

```python
new.to_csv('new_data.csv',index=False)
```

### Extracting features of 2018 movies from Wikipedia

```python
import pandas as pd
import numpy as np
```

```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2018"
df1 = pd.read_html(link, header=0)[2]
df2 = pd.read_html(link, header=0)[3]
df3 = pd.read_html(link, header=0)[4]
df4 = pd.read_html(link, header=0)[5]
```

```python
df = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
```

```python
df
```

```python
from omdbapi.movie_search import GetMovie
import json
import requests
```

```python
genres = []
for i in range(268):
    movie = GetMovie(title=df['Title'][i], api_key='e3fc1d3f', plot='full')
    data = movie.get_data('Genre')
    genres.append(data)
```

```python
from pandas import DataFrame
df3 = DataFrame(genres,columns=['Genre'])
```

```python
df['genres'] = df3
```

```python
df_en = df[['Title','Cast and crew','genres']]
```

```python
df_en
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
df_en['director_name'] = df_en['Cast and crew'].map(lambda x: get_director(x))
```

```python
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python
df_en['actor_1_name'] = df_en['Cast and crew'].map(lambda x: get_actor1(x))
```

```python
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
df_en['actor_2_name'] = df_en['Cast and crew'].map(lambda x: get_actor2(x))
```

```python
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
df_en['actor_3_name'] = df_en['Cast and crew'].map(lambda x: get_actor3(x))
```

```python
df_en = df_en.rename(columns={'Title':'movie_title'})
```

```python
new_df18 = df_en.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
new_df18['actor_2_name'] = new_df18['actor_2_name'].replace(np.nan, 'unknown')
new_df18['actor_3_name'] = new_df18['actor_3_name'].replace(np.nan, 'unknown')
new_df18['movie_title'] = new_df18['movie_title'].str.lower()
```

```python
new_df18['comb'] = new_df18['actor_1_name'] + ' ' + new_df18['actor_2_name'] + ' '+ new_df18['actor_3_name'] + ' '+ new_df18['director_name'] +' ' + new_df18['genres']
```
